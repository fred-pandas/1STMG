```python
from random import random
import pylab as pl
```

```python
def pile(p):
    if random() <= p:
        return 1
    else:
        return 0
```

```python
def frequences(p):
    c = 0
    for i in range(1000):
        if pile(p) + pile(p) + pile(p) == 2:
            c = c + 1
    return c/1000
```

```python
frequences(0.7)
```

```python
F = [frequences(0.7) for i in range(2000)]
pl.figure(figsize=(11,3))
pl.subplot(121)
pl.plot(F,'.')
pl.subplot(122)
pl.hist(F,bins=20,orientation='horizontal',edgecolor='black')
pl.show()
```
